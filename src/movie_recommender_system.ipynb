```python
# it is a list of dixtionaries 
# we need to convert its format into ['Action','Adventure','Fantasy','Science Fiction']
```

```python
def convert_format(obj):
    ls = []
    for i in ast.literal_eval(obj):
        ls.append(i['name'])
    return ls
```

```python
# convert_format('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
# it is in string format we need to change into a list
# error-> string indices must be integers, not 'str'
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
df['genres'] = df['genres'].apply(convert_format)
```

```python
df.head(1)
```

```python
df['keywords'] = df['keywords'].apply(convert_format)
```

```python
# I only want name of top 3 working actors for each movie
```

```python
def convert_format2(obj):
    ls = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            ls.append(i['name'])
            count+=1
        else:
            break
    return ls
```

```python
df['cast'] = df['cast'].apply(convert_format2)
```

```python
df.head(1)
```

```python
df['crew'][0]
```

```python
# i want only those name which are working as director
def convert_format3(obj):
    ls = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            ls.append(i['name'])
            break
    return ls
```

```python
df['crew'] = df['crew'].apply(convert_format3)
```

```python
df.head(2)
```

```python
# overview col is a string convert it into a list
df['overview'][0]
```

```python
ls = (df['overview'][0]).split()
ls
```

```python
df['overview'] = df['overview'].apply(lambda x:x.split())
```

```python
df.head(2)
```

## we need to remove any space between a tag like (science fiction as sciencefiction) or (sam mendes as sammendes)

```python
df['genres'] = df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df['cast'] = df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df['keywords'] = df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df['crew'] = df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

## now our last job is to concatenate all the cols in single col as tags

```python
df['tags'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
```

```python
df.head(2)
```

```python
final_df = df[['movie_id','title','tags']]
```

```python
final_df.head(1)
```

```python
# convert tags into string
```

```python
final_df['tags'] = final_df['tags'].apply(lambda x:" ".join(x))
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
final_df['tags'] = final_df['tags'].apply(stem)
```

```python
final_df['tags'][0]
```

```python
# convert every alphabet into lower case --> recommended
```

```python
final_df['tags'] = final_df['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = 'english')
```

```python
vectors = cv.fit_transform(final_df['tags']).toarray()
```

```python
cv.fit_transform(final_df['tags']).toarray().shape
```

```python
vectors
```

```python
print(cv.get_feature_names_out())
```

```python
# stemming -> reducing words to there base forms
# loves,loved,loving -> love,love,love
ps.stem('loved')
```

```python
# measure cosine distance between each and every vector to look for similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
def recommend(movie):
    movie_index = final_df[final_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x:x[1])[1:6]
    for i in movies_list:
        print(final_df.iloc[i[0]].title)
```
